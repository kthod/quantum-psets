<h2> Έρώτημα 1</h2>

```python
from qiskit import QuantumCircuit,QuantumRegister,assemble,Aer
import numpy as np

from math import sqrt, pi
from qiskit.visualization import  plot_bloch_multivector
```

```python
x = QuantumRegister(1,'u1')
y = QuantumRegister(1, 'u2')


qc = QuantumCircuit(x,y)

qc.initialize([1/sqrt(2),1/sqrt(2)],0)
qc.initialize([1/sqrt(2),-1/sqrt(2)],1)
qc.x(y)
qc.x(x)

qc.draw(output='mpl')

```

```python
svsim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
final_state = svsim.run(qobj).result().get_statevector()
plot_bloch_multivector(final_state,title ="bisias")
```
